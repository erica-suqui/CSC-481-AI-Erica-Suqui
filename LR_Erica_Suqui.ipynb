```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('student scores.csv')
df.head()
```

```python
df.shape
```

```python
x = df.drop('Scores',axis=1)
y = df['Scores']
```

```python
y
```

```python
plt.figure(figsize=(4,4))


plt.scatter(x,y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)

plt.show
```

```python
from sklearn import datasets, model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error




```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=12)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)


```

```python
alpha= model.intercept_
beta = model.coef_

print("Alpha","{:.2f}".format(alpha))
print("beta",beta)

print(f"y={alpha}+{beta}X")
```

```python
y_pred_train= model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
plt.figure(figsize=(7,3))


#subplot 1
plt.subplot(1,2,1)
plt.scatter(y_train,y_pred_train)
plt.title("Training case")
plt.xlabel('Y Train ')
plt.ylabel('Y Train Predict')
plt.grid(True)

#subplot 1
plt.subplot(1,2,2)
plt.scatter(y_test,y_pred_test)
plt.title("Testing case")
plt.xlabel('Y_Test Train')
plt.ylabel('Y_Test Predict')
plt.grid(True)

plt.show()
```

```python
mse_train= mean_squared_error(y_train,y_pred_train)
mae_train= mean_absolute_error(y_train,y_pred_train)


mse_test= mean_squared_error(y_test,y_pred_test)
mae_test= mean_absolute_error(y_test,y_pred_test)

```

```python
#create a dataframe of the results


performance_df = pd.DataFrame({
    'Metrics': ['MSE','MAE','MSE','MAE'],
    'DataType':['train','test','trian','test'],
    'Values':[mse_train,mae_train,mse_test,mae_test]
})
performance_df.T
```
