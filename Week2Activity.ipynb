```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,15,7,12,8]


#create a line plot

plt.plot(x,y, marker ='o',markersize=8,color='blue',linestyle='-.',linewidth=4)

plt.grid(True
         )
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('x versus y')



plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year':[2010,2011,2012,2013,2014],
    'Sales':[500,800,1700,1500,2000]
}

df = pd.DataFrame(data)


plt.plot(df['Year'],df['Sales'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Salaes Over Years')
plt.grid(True)


#Set x-axis ticks at integer values
#plt.xticks(df['Year'])
plt.show()
```

```python
data = {
    'Category':['A','B','C','D'],
    'Sales':[500,800,1200,1500]
}

df = pd.DataFrame(data)

plt.scatter(df['Category'],df['Sales'],color = 'red')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Create a dataframe

data = {
    'Age':[25,30,22,35,28],
    'Income':[5000,60000,4500,70000,55000]
}


df = pd.DataFrame(data)

#Create a Scatterplot
plt.scatter(df['Age'],df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income vs Age')
plt.grid('True')
plt.show()
```

```python
#create a dataframe

data = {
    'Scores':[85,78,92,65,80,75,88,90,82,70]
}

df = pd.DataFrame(data)



plt.hist(df['Scores'],width=1.5)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('distribution of Scores')
plt.grid(True)
plt.show()
```
